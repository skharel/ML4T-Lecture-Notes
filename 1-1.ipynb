# Read columns and gather statistics

```python
def get_max_close (symbol):
    """Return the maximum closing value for stock indicated by symbol.

    Note: Data for a stock is stored in file: data/<symbol>.csv
    """
    df = pd.read_csv("data/{}.csv".format(symbol))  # read in data
    return  df['Close'].max()
```

```python
aaplMaxClose = get_max_close('AAPL')
ibmMaxClose = get_max_close('IBM')
print(f"AAPL {aaplMaxClose}; IBM {ibmMaxClose}")
```

```python
def get_mean_volume(symbol):
    df = pd.read_csv("data/{}.csv".format(symbol))  # read in data
    return  df['Volume'].mean()
```

```python
aaplMean = get_mean_volume('AAPL')
ibmMean = get_mean_volume('IBM')
print(f"AAPL {aaplMean}; IBM {ibmMean}")
```

```python
def get_column(symbol, col):
    df = pd.read_csv(f"data/{symbol}.csv")
    return df[col]
```

# Plot a column

```python
aaplAdjClose = get_column('AAPL', 'Adj Close')
aaplAdjClose.plot()
plt.show()
```

```python
ibmHigh = get_column('IBM', 'High')
ibmHigh.plot()
```
